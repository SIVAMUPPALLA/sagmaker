```python
import pandas as pd
data=pd.read_csv("oasis_new.csv")
data.head()
```

```python
dframe=data.drop(["Subject ID","MRI ID","Hand","id","EventTime"], axis=1)
dframe.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder(handle_unknown='ignore').fit(dframe[["Visit","M/F"]])
```

```python
oh
```

```python
oh.get_feature_names(["Visit","M/F"])
```

```python
one_hot_data=oh.transform(dframe[["Visit","M/F"]])
one_hot_data
```

```python
one_hot_df=pd.DataFrame(one_hot_data.toarray(),columns=oh.get_feature_names(["Visit","M/F"]))
```

```python
one_hot_df.head()
```

```python
dframe[one_hot_df.columns]=one_hot_df
```

```python
dframe.head()
```

### Drop the cloumn

```python
dframe=dframe.rename(columns={"M/F_F":"female","M/F_M":"male" })
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
dframe["Group"]=LE.fit_transform(dframe["Group"])
```

```python
dframe.head()
```

```python
dframe.drop(["Visit","M/F"], axis=1,inplace=True)
```

```python
dframe.head()
```

```python
from collections import Counter
Counter(dframe["Group"])
```

```python
LE.inverse_transform([0,1,2])
```

```python
dframe=dframe[dframe["Group"]!=0]
```

```python
dframe.head()
```

```python
Counter(dframe["Group"])
```

```python
dframe.fillna(0,inplace=True)
```

```python
dframe.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(dframe, test_size=0.2, random_state=41)
train,valid=train_test_split(train,test_size=0.25, random_state=41)
```

```python
train.shape, test.shape, valid.shape
```

```python
train.to_csv("siva_train.csv", index=False, header=False)
valid.to_csv("siva_valid.csv", index=False, header=False)
test.to_csv("siva_test.csv", index=False, header=False)
```
